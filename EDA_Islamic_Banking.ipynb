## 1. Import Libraries

# Learning Objective
Import the required Python libraries for data analysis and visualization.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"]=(10,5)
pd.set_option("display.max_columns",None)
```

## 2. Load Dataset

# Learning Objective
Read the Excel dataset and verify that it loads correctly.

> Update the file path if necessary.


**Expected Outcome:** Students should be able to explain the results before moving to the next section.

```
file_path="ISLAMIC_BANKING_DATA_202606011047.xlsx"

df=pd.read_excel(file_path,header=7)
df.head()
```

## 3. Understand the Dataset

Inspect the overall structure of the dataset.

**Expected Outcome:** Students should be able to explain the results before moving to the next section.

```
print("Shape:",df.shape)
display(df.head())
display(df.tail())
display(df.info())
display(df.describe(include='all'))
```

## 4. Data Quality Assessment

Check missing values, duplicates and data types.

**Expected Outcome:** Students should be able to explain the results before moving to the next section.

```
missing=df.isnull().sum().sort_values(ascending=False)
display(missing[missing>0])

print("Duplicate rows:",df.duplicated().sum())
display(df.dtypes)
```

## 5. Univariate Analysis

Explore categorical and numerical variables.

**Expected Outcome:** Students should be able to explain the results before moving to the next section.

```
cat_cols=df.select_dtypes(include='object').columns
num_cols=df.select_dtypes(include=np.number).columns

for c in cat_cols:
    print("\n",c)
    display(df[c].value_counts().head(10))

display(df[num_cols].describe().T)
```

## 6. Visualizations

Create introductory charts.

**Expected Outcome:** Students should be able to explain the results before moving to the next section.

```
if 'Country' in df.columns:
    df['Country'].value_counts().head(10).plot(kind='bar')
    plt.title("Top 10 Countries")
    plt.show()

if 'Values in USD Millions' in df.columns:
    sns.histplot(df['Values in USD Millions'].dropna(),kde=True)
    plt.title("Distribution of Values in USD Millions")
    plt.show()
```

## 7. Time Analysis

Analyse the reporting period.

**Expected Outcome:** Students should be able to explain the results before moving to the next section.

```
if 'Time Period' in df.columns:
    display(df['Time Period'].value_counts().sort_index())
```

## 8. Correlation

Investigate relationships among numeric variables.

**Expected Outcome:** Students should be able to explain the results before moving to the next section.

```
corr=df.select_dtypes(include=np.number).corr(numeric_only=True)
display(corr)

sns.heatmap(corr,annot=True,cmap='Blues')
plt.title("Correlation Matrix")
plt.show()
```
