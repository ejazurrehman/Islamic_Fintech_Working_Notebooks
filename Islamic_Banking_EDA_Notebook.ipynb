# Step 06: Data Cleaning & Data Quality

> Athough `isnull()` reports no missing values, inspection shows that some variables contain blank strings `(" ")`, which represent hidden missing values. These should be converted to `NaN` before further analysis.

```python
# Replace blank strings with NaN
df = df.replace(r'^\s*$', np.nan, regex=True)
```

```python
# Check for Missing Values Again
df.isnull().sum()
```

> The data cleaning process successfully replaced blank strings with NaN values and converted the financial variables (Values and Values in USD Millions) to the float64 data type. The reduced non-null counts confirm that hidden missing values were identified and properly handled. The dataset is now ready for descriptive statistics, visualization, and further numerical analysis.

```python
variation = pd.DataFrame({
    "Mean": df[numeric_cols].mean(),
    "Std Dev": df[numeric_cols].std()
})

variation["Coefficient of Variation"] = (
    variation["Std Dev"] /
    variation["Mean"]
)

variation.sort_values(
    "Coefficient of Variation",
    ascending=False
)
```

Understanding the Coefficient of Variation (CV)
> The Coefficient of Variation (CV) measures the relative variability of a variable by comparing its standard deviation to its mean. A higher CV indicates greater variability relative to the average value.

Formula:

$$
\text{Coefficient of Variation (CV)} = \frac{\text{Standard Deviation}}{\text{Mean}}
$$

# Step 11: Correlation

Definition
> Correlation measures the strength and direction of the relationship between two numerical variables. The correlation coefficient ranges from −1 to +1.

Why are we doing it?
> Correlation analysis helps identify variables that move together, detect redundant variables, and understand relationships before performing statistical modelling or machine learning.

Formula
$$
r=\frac{\mathrm{Cov}(X,Y)}{\sigma_X\sigma_Y}
$$

Where:

- $r$ = Correlation coefficient
- $\mathrm{Cov}(X,Y)$ = Covariance between two variables
- $\sigma_X, \sigma_Y$ = Standard deviations of $X$ and $Y$

## Research Questions
- Which numerical variables are strongly related?
- Can any variables be considered redundant?

### Student Interpretation

Answer-01
> Examine the heatmap for correlation coefficients close to +1 or −1. Variables with high positive or negative correlation have a strong linear relationship and tend to move together.

Answer-02
> Variables with very high positive correlation (e.g., r>0.90) may contain similar information and could be considered redundant in predictive modelling. Such variables should be evaluated before feature selection.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation matrix
corr = df.select_dtypes(include="number").corr()

# Short labels
labels = ["USD Rate", "Values", "USD Millions"]

plt.figure(figsize=(7,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="RdYlBu_r",
    fmt=".2f",
    square=True,
    linewidths=0.5,
    xticklabels=labels,
    yticklabels=labels,
    cbar_kws={"shrink":0.8},
    annot_kws={"fontsize":11}
)

plt.title(
    "Correlation Matrix",
    fontsize=16,
    fontweight="bold"
)

plt.xticks(rotation=0, fontsize=11)
plt.yticks(rotation=0, fontsize=11)

plt.tight_layout()
plt.show()
```
