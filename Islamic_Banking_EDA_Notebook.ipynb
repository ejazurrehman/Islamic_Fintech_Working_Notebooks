# RQ4: Does the dataset contain missing values or data quality issues?

A conventional missing-value check (isnull()) reports no missing values because every cell contains a value.

However, the dataset uses blank spaces (" ") instead of true NaN values in many fields.

Examples include:

- Values
- Values in USD Millions
- Currency
- Units

These blank strings represent unavailable or non-applicable information and should be converted to NaN during preprocessing.

Other observations include:

- No duplicate column names
- Consistent quarterly time format
- Consistent country names
- Well-defined indicator codes
- Mixed numeric and text values due to blank strings

Therefore, the primary data quality issue is the presence of blank-string placeholders rather than actual missing values.

# Step 3: Missing Value Analysis

Missing values are common in financial datasets.

A missing value does not always indicate an error. In many cases, financial institutions may not report certain indicators because:

- the indicator is not applicable,
- reporting requirements differ,
- data collection is incomplete.

Before removing missing values, analysts must understand why they exist.

---

## Research Question

RQ8

Which variables contain missing values, and what are the possible reasons?

```python
# Missing Values

missing = pd.DataFrame(df.isnull().sum(), columns=["Missing Values"])

missing["Percentage"] = (
    missing["Missing Values"] / len(df) * 100
)

missing.sort_values(
    by="Percentage",
    ascending=False
)
```

```python
# Missing Value Visualization

plt.figure(figsize=(8,6))

sns.heatmap(
    df.isnull(),
    cbar=False,
    cmap="viridis"
)

plt.title("Missing Value Heatmap")

plt.show()
```
