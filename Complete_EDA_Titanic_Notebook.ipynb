# Part A: Project Foundation

## 1. What Is Exploratory Data Analysis?

>**Exploratory Data Analysis (EDA)** is the systematic process of examining, summarizing, visualizing, and questioning data before formal modelling or decision-making.

EDA helps an analyst answer questions such as:

- What does the dataset represent?
- What is the unit of analysis?
- Which variables are available?
- What types of variables are present?
- Is the data complete?
- Are there unusual or impossible values?
- What patterns appear in individual variables?
- Which variables seem related?
- Are important relationships hidden by other variables?
- What additional variables can be created?
- Which findings are strong enough to support a conclusion?

EDA is not simply the production of charts. A chart without a question and interpretation is only a graphic.

> **EDA is a reasoning process supported by statistics, Python, and visualization.**

## 2. Why EDA Comes Before Modelling

>A predictive model can process data without understanding the real-world meaning of the variables. An analyst cannot.

If we move directly to modelling, we may:

- use the wrong data type,
- ignore missing values,
- retain duplicate records,
- misunderstand categorical codes,
- allow outliers to distort results,
- create data leakage,
- misinterpret correlation,
- or produce a technically correct model based on analytically weak data.

Therefore, the analytical sequence should be:

### Step 1: Understand the problem  
### Step 2: Understand the data  
### Step 3: Assess data quality  
### Step 4: Explore individual variables  
### Step 5: Explore relationships  
### Step 6: Engineer useful features  
### Step 7: Summarize evidence  
### Step 8: Only then consider modelling

This notebook focuses on Steps 1 to 7.

## 3. The EDA Mindset

A strong analyst does not begin with:

> "Which chart should I make?"

A strong analyst begins with:

> "What do I need to understand?"

The chart, statistical measure, table, or Python method is selected **after** the analytical question is clear.

### Example

Weak approach:

```python
sns.countplot(data=df, x="Pclass")
```

Analytical approach:

**Question:** How were passengers distributed across travel classes?

**Method:** Count the number and percentage of passengers in each class.

**Visualization:** Use a count plot because `Pclass` is a categorical or ordinal variable.

**Interpretation:** Explain whether the passenger population was concentrated in a particular class.

**Connection:** Investigate whether survival outcomes differed across those classes.

The difference is not the code. The difference is the reasoning.

### Important Definition:

1. Categorical Variable
   
Definition

>A categorical variable places observations into different groups or categories. Its values describe what type or group an observation belongs to rather than measuring a numerical quantity.

Simple Example

Suppose we have a variable called Payment_Method:

| Customer | Payment_Method |
| -------- | -------------- |
| 1        | Cash           |
| 2        | Card           |
| 3        | Bank Transfer  |
| 4        | Cash           |

Titanic Example:

Sex\
Possible values:

male\
female

>Sex is a categorical variable because it places passengers into categories.

Another example is:

Embarked
| Code | Meaning     |
| ---- | ----------- |
| C    | Cherbourg   |
| Q    | Queenstown  |
| S    | Southampton |

Key Point

>A categorical variable answers: “Which group or category?”

2. Identifier-Like Text
   
Definition

>Identifier-like text is a text variable mainly used to identify a person, object, account, transaction, or record.

Its values may contain:

- letters,
- numbers,
- symbols,
or combinations of them.

>Although the values may repeat or contain patterns, they are not ordinary categories such as male/female or city names.

Simple Example

Consider transaction codes:

| Transaction | Transaction_ID |
| ----------- | -------------- |
| 1           | TXN-1001       |
| 2           | TXN-1002       |
| 3           | TXN-1003       |

It would not make sense to calculate:

- the mean Transaction ID,
- the median Transaction ID,
or a histogram of Transaction IDs.

The values mainly identify records.

Titanic Example
Ticket

Example values:

A/5 21171
PC 17599
STON/O2. 3101282
113803

Ticket is called identifier-like text because:

- it contains letters and numbers,
- it identifies ticket records,
- some passengers may share the same ticket,
and patterns inside the ticket may later provide useful information.

Important Difference

>PassengerId is a pure identifier because each passenger has a unique ID.

>Ticket is identifier-like text because it behaves partly like an identifier, but repeated ticket values may reveal passenger groups.

Key Point

>Identifier-like text answers: “Which record, code, account, or reference?”

3. Binary Categorical Target

This term contains three separate ideas:

Binary

>Binary means there are only two possible categories.

Examples:

| Variable       | Categories                 |
| -------------- | -------------------------- |
| Loan Default   | Yes / No                   |
| Fraud          | Fraud / Not Fraud          |
| Customer Churn | Churn / No Churn           |
| Survival       | Survived / Did Not Survive |

Categorical:

>The two values represent categories rather than measured quantities.

Target:

>A target variable is the main outcome we want to explain, investigate, or predict.

Titanic Example
Survived

| Value | Meaning         |
| ----- | --------------- |
| 0     | Did not survive |
| 1     | Survived        |

Therefore, Survived is a binary categorical target because:

- Binary: it has only two outcomes, 0 and 1
- Categorical: these values represent outcome categories
- Target: survival is the main outcome we want to investigate

Important Warning

Although Python stores Survived as numbers:

0
1

we should not interpret them as ordinary numerical measurements.

For example:

1 is not "twice as much survival" as 0.

The numbers are simply codes for two categories.

Key Point

>A binary categorical target answers: “Which of the two possible outcomes occurred?”

4. Ordinal Categorical Variable
   
Definition

>An ordinal categorical variable contains categories that have a meaningful order or ranking.

However, the numerical distance between categories is not necessarily equal or measurable.

Simple Example

Consider customer satisfaction:

| Code | Satisfaction Level |
| ---- | ------------------ |
| 1    | Low                |
| 2    | Medium             |
| 3    | High               |

There is a clear order:

Low < Medium < High

But we cannot automatically say:

High satisfaction is exactly three times Low satisfaction.

The numbers represent rank, not exact quantity.

Titanic Example
Pclass

| Value | Passenger Class |
| ----- | --------------- |
| 1     | First Class     |
| 2     | Second Class    |
| 3     | Third Class     |

There is a meaningful order based on travel class:

>First Class → Second Class → Third Class

Therefore, Pclass is an ordinal categorical variable.

Important Warning

Python stores Pclass as:

int64

But analytically, it is not an ordinary continuous numerical variable.

For example, we should not say:

“Third class is exactly three times first class.”

That interpretation would be meaningless.

Key Point

An ordinal categorical variable answers: “Which ordered level or rank?”

# Part J: Outlier Analysis

## 56. What Is an Outlier?

An outlier is an observation that lies unusually far from the main body of data.

An outlier may be:

- a data-entry error,
- a measurement error,
- a rare but valid observation,
- or an important real-world case.

### Critical Rule

> **Detection is not deletion.**

We first identify unusual values, then interpret them in context.

## 57. IQR Method

The Interquartile Range is:

\[
IQR = Q3 - Q1
\]

Common boundaries are:

\[
Lower = Q1 - 1.5(IQR)
\]

\[
Upper = Q3 + 1.5(IQR)
\]

````python
def iqr_outlier_report(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr

    outliers = data[(data[column] < lower) | (data[column] > upper)]

    return {
        "Variable": column,
        "Q1": q1,
        "Q3": q3,
        "IQR": iqr,
        "Lower_Bound": lower,
        "Upper_Bound": upper,
        "Outlier_Count": len(outliers),
        "Outlier_Percent": len(outliers) / len(data) * 100
    }

pd.DataFrame([
    iqr_outlier_report(df_eda, "Age"),
    iqr_outlier_report(df_eda, "Fare")
])
````

# Part M: Correlation Analysis

## 68. What Is Correlation?

Correlation measures the direction and strength of a linear relationship between numerical variables.

The Pearson correlation coefficient ranges from:

\[
-1 \leq r \leq 1
\]

- near `+1`: strong positive linear relationship
- near `-1`: strong negative linear relationship
- near `0`: weak linear relationship

### Important Limitations

- correlation does not prove causation,
- low correlation does not mean no relationship,
- categorical variables require careful interpretation,
- nonlinear relationships may be missed.

# Part N: Feature Engineering

## 71. What Is Feature Engineering?

Feature engineering creates new variables from existing information.

In EDA, engineered features can reveal patterns hidden in raw variables.

We will create:

- `FamilySize`
- `IsAlone`
- `Title`
- `Deck`

These features must have a clear analytical rationale.

## 72. Create Family Size

`SibSp` counts siblings and spouses.  
`Parch` counts parents and children.

A passenger's total family group aboard can be approximated as:

\[
FamilySize = SibSp + Parch + 1
\]

The `+1` includes the passenger.

````python
df_eda["FamilySize"] = df_eda["SibSp"] + df_eda["Parch"] + 1

df_eda["FamilySize"].describe()
````

````python
family_survival = (
    df_eda.groupby("FamilySize")["Survived"]
    .agg(["count", "mean"])
    .rename(columns={"mean": "Survival_Rate"})
)

family_survival["Survival_Rate"] *= 100
family_survival
````

## 73. Create `IsAlone`

````python
df_eda["IsAlone"] = (df_eda["FamilySize"] == 1).astype(int)

df_eda["IsAlone"].value_counts()
````

````python
alone_survival = (
    df_eda.groupby("IsAlone")["Survived"]
    .agg(["count", "mean"])
    .rename(columns={"mean": "Survival_Rate"})
)

alone_survival["Survival_Rate"] *= 100
alone_survival
````

## 74. Extract Passenger Title from Name

Names contain structured information such as:

- Mr
- Mrs
- Miss
- Master
- rare professional or social titles

This may help explain age and social grouping.

````python
df_eda["Title"] = df_eda["Name"].str.extract(r",\s*([^.]*)\.")

df_eda["Title"].value_counts()
````

## 75. Consolidate Rare Titles

````python
title_counts = df_eda["Title"].value_counts()

rare_titles = title_counts[title_counts < 10].index

df_eda["Title_Grouped"] = df_eda["Title"].replace(rare_titles, "Rare")

df_eda["Title_Grouped"].value_counts()
````

## 76. Revisit Missing Age Using Title

We previously delayed age imputation because passenger age may differ systematically across groups.

Now compare age by title.

````python
df_eda.groupby("Title_Grouped")["Age"].agg(
    ["count", "mean", "median"]
).sort_values("median")
````

### Better Missing-Age Decision

A title-based median can be more defensible than one overall mean because titles capture meaningful age differences.

We preserve the original `Age` and create a separate prepared variable.

````python
df_eda["Age_Filled"] = df_eda["Age"]

title_medians = df_eda.groupby("Title_Grouped")["Age"].transform("median")

df_eda["Age_Filled"] = df_eda["Age_Filled"].fillna(title_medians)

print("Missing Age before:", df_eda["Age"].isna().sum())
print("Missing Age_Filled after:", df_eda["Age_Filled"].isna().sum())
````

## 77. Extract Deck from Cabin

The first letter of a cabin number can represent the deck.

Missing cabin values will be labelled `Unknown`.

````python
df_eda["Deck"] = df_eda["Cabin"].str[0].fillna("Unknown")

df_eda["Deck"].value_counts()
````

## 78. Handle Missing Embarked Values

Because missingness is very small, mode imputation is a simple documented choice.

We preserve the original variable and create `Embarked_Filled`.

````python
df_eda["Embarked_Filled"] = df_eda["Embarked"].fillna(
    df_eda["Embarked"].mode()[0]
)

print("Missing Embarked before:", df_eda["Embarked"].isna().sum())
print("Missing Embarked_Filled after:", df_eda["Embarked_Filled"].isna().sum())
````

### Feature Engineering Decision

We have not overwritten the original variables. This preserves transparency.

> **Connection to Part O:**  
> We can now integrate original and engineered variables to answer deeper analytical questions.

### Instructor Answer: Feature Engineering

`FamilySize` combines the two separate family-count variables into one more interpretable measure. `IsAlone` identifies passengers with no recorded family member aboard.

Passenger titles reveal strong age differences. After title-based median imputation, missing values in `Age_Filled` fall from **177 to 0**.

`Deck` preserves partial information from the highly incomplete `Cabin` variable without pretending that missing cabin numbers are known.

`Embarked_Filled` removes the remaining **2** missing embarkation values using the modal category.

**Decision:** Keep both original and prepared variables so the analytical process remains transparent and auditable.
