### 6. Floor Division Operator (//)
- The floor division operator (//) divides one number by another and returns the quotient without the decimal part. It rounds the result down to the nearest whole number (toward negative infinity).

### 7. Exponentiation Operator (**)

### 8. BODMAS
- it is a rule that determines the correct order of mathematical operations in an expression.

| Letter | Meaning        | Description                                                        |
| ------ | -------------- | ------------------------------------------------------------------ |
| **B**  | Brackets       | Solve expressions inside brackets first: `()`, `[]`, `{}`          |
| **O**  | Orders         | Evaluate powers (exponents), roots, and indices, e.g., `2³`, `√16` |
| **D**  | Division       | Perform division from left to right                                |
| **M**  | Multiplication | Perform multiplication from left to right                          |
| **A**  | Addition       | Perform addition from left to right                                |
| **S**  | Subtraction    | Perform subtraction from left to right                             |


### 9. PEMDAS
- It is another rule for the order of mathematical operations, primarily used in the United States. It is equivalent to BODMAS, which is commonly used in the UK, Pakistan, and many Commonwealth countries.

| Letter | Meaning        | Description                                          |
| ------ | -------------- | ---------------------------------------------------- |
| **P**  | Parentheses    | Solve expressions inside `()`, `[]`, and `{}` first. |
| **E**  | Exponents      | Evaluate powers (e.g., `2³`, `5²`) and roots.        |
| **M**  | Multiplication | Perform multiplication from left to right.           |
| **D**  | Division       | Perform division from left to right.                 |
| **A**  | Addition       | Perform addition from left to right.                 |
| **S**  | Subtraction    | Perform subtraction from left to right.              |

### Simple Interest Calculation

```python
principal = float(input("Enter Principal Amount (Rs.): "))
rate = float(input("Enter Interest Rate (%): "))
time = int(input("Enter Time (Years): "))

interest = (principal * rate * time) / 100

print("Simple Interest = Rs.", interest)
```

### Profit or Loss (Using if and else)

```python
revenue = float(input("Enter Revenue (Rs.): "))
expenses = float(input("Enter Expenses (Rs.): "))

if revenue > expenses:
    print("The company earned a Profit.")
else:
    print("The company incurred a Loss.")
```

### Investment Recommendation (Using if-else)

```python
expected_return = float(input("Enter Expected Return (%): "))

if expected_return >= 15:
    print("Good Investment")
else:
    print("Investment Needs Further Analysis")
```

### Credit Rating (Using if-elif-else)

```python
credit_score = int(input("Enter Credit Score: "))

if credit_score >= 750:
    print("Excellent Credit")
elif credit_score >= 650:
    print("Good Credit")
else:
    print("Poor Credit")
```

### Loan Approval System

```python
 # Function to evaluate loan application
def loan_approval():

    # Input Variables
    applicant_name = input("Enter Applicant Name: ")
    monthly_income = float(input("Enter Monthly Income (Rs.): "))
    credit_score = int(input("Enter Credit Score: "))
    loan_amount = float(input("Enter Loan Amount (Rs.): "))

    print("\n----- Loan Application Summary -----")
    print("Applicant Name :", applicant_name)
    print("Monthly Income :", monthly_income)
    print("Credit Score   :", credit_score)
    print("Loan Amount    :", loan_amount)

    # Conditional Logic
    if monthly_income >= 100000 and credit_score >= 750:
        print("\nLoan Status : Approved")
        print("Reason : High income and excellent credit score.")

    elif monthly_income >= 70000 and credit_score >= 650:
        print("\nLoan Status : Conditional Approval")
        print("Reason : Requires additional documents.")

    elif monthly_income < 50000 or credit_score < 600:
        print("\nLoan Status : Rejected")
        print("Reason : Low income or poor credit score.")

    else:
        print("\nLoan Status : Under Review")
        print("Reason : Further financial assessment required.")

# Function Call
loan_approval()
```

# Loops in Python

Definition

A loop is a programming structure that repeatedly executes a block of code until a specified condition is met or for a specified number of iterations. Loops eliminate the need to write the same code multiple times.

Python provides two main types of loops:

1. for loop: Used when the number of repetitions is known.
2. while loop: Used when the number of repetitions depends on a condition.

1. for Loop

Definition: A for loop executes a block of code a fixed number of times or iterates over a sequence such as a list, range, or string.

Syntax:\
for variable in sequence:\
    # Code to execute

### Calculate Total Investment of 5 Investors

```python
total_investment = 0

for i in range(1, 6):
    investment = float(input(f"Enter Investment of Investor {i} (Rs.): "))
    total_investment += investment

print("\nTotal Investment = Rs.", total_investment)
```

### Calculate Annual Profits

```python
profits = [120000, 145000, 98000, 175000]

for profit in profits:
    print("Annual Profit = Rs.", profit)
```

### Calculate Zakat for Multiple Customers

```python
nisab = 180000

for i in range(1, 6):

    savings = float(input(f"Customer {i} Savings (Rs.): "))

    if savings >= nisab:
        zakat = savings * 0.025
        print("Zakat Payable = Rs.", zakat)

    else:
        print("Not Eligible for Zakat")
```

### Screen Multiple Companies

```python
for i in range(1, 4):

    company = input("Company Name: ")
    debt_ratio = float(input("Debt Ratio (%): "))

    if debt_ratio <= 33:
        print(company, "is Shariah-Compliant")

    else:
        print(company, "is Non-Compliant")
```
