# Step 3: Missing Value Analysis

Missing values are common in financial datasets.

A missing value does not always indicate an error. In many cases, financial institutions may not report certain indicators because:

- the indicator is not applicable,
- reporting requirements differ,
- data collection is incomplete.

Before removing missing values, analysts must understand why they exist.

---

## Research Question

## RQ8: Which variables contain missing values, and what are the possible reasons?

```python
# Missing Values

missing = pd.DataFrame(df.isnull().sum(), columns=["Missing Values"])

missing["Percentage"] = (
    missing["Missing Values"] / len(df) * 100
)

missing.sort_values(
    by="Percentage",
    ascending=False
)
```

```python
# Missing Value Visualization

plt.figure(figsize=(8,6))

sns.heatmap(
    df.isnull(),
    cbar=False,
    cmap="viridis"
)

plt.title("Missing Value Heatmap")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Count observations by country
country_counts = df["Country"].value_counts()

# Country-inspired colors
colors = {
    "Pakistan": "#0B6623",      # Dark Green
    "Indonesia": "#D50000",     # Red
    "Malaysia": "#0A2F73",      # Navy Blue
    "Saudi Arabia": "#006C35",  # Saudi Green
    "Bangladesh": "#006A4E"     # Bangladesh Green
}

bar_colors = [colors[country] for country in country_counts.index]

# Create figure
plt.figure(figsize=(10, 6))

bars = plt.bar(
    country_counts.index,
    country_counts.values,
    color=bar_colors,
    edgecolor="black",
    linewidth=1.2
)

# Add value labels
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height() + 150,
        f"{int(bar.get_height()):,}",
        ha="center",
        va="bottom",
        fontsize=11,
        fontweight="bold"
    )

# Titles and labels
plt.title("Observations by Country", fontsize=16, fontweight="bold")
plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Observations", fontsize=12)

# Formatting
plt.xticks(rotation=0, fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis="y", linestyle="--", alpha=0.4)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Top 20 indicator descriptions
indicator_counts = df["Indicator Description"].value_counts().head(20)

# Create figure
plt.figure(figsize=(12, 8))

# Modern color palette
colors = plt.cm.viridis_r(
    [i / len(indicator_counts) for i in range(len(indicator_counts))]
)

# Horizontal bars
bars = plt.barh(
    indicator_counts.index,
    indicator_counts.values,
    color=colors,
    edgecolor="black",
    linewidth=0.8
)

# Add values to bars
for bar in bars:
    width = bar.get_width()
    plt.text(
        width + 20,
        bar.get_y() + bar.get_height()/2,
        f"{int(width):,}",
        va="center",
        fontsize=10,
        fontweight="bold"
    )

# Titles
plt.title(
    "Top 20 Most Frequently Reported Financial Indicators",
    fontsize=16,
    fontweight="bold",
    pad=15
)

plt.xlabel("Number of Observations", fontsize=12)
plt.ylabel("Indicator Description", fontsize=12)

# Grid
plt.grid(axis="x", linestyle="--", alpha=0.4)

# Format x-axis
plt.gca().xaxis.set_major_formatter(
    mtick.StrMethodFormatter('{x:,.0f}')
)

# Highest value at top
plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

# Step 09: Time Analysis

Definition
> Time Analysis examines how observations and financial indicators change across different reporting periods.

Why are we doing it?
> It helps identify trends, reporting consistency, seasonal patterns, and changes in Islamic Banking performance over time.

## Research Questions

- Is reporting consistent over time?
- Are there missing quarters?
- Which periods contain the greatest reporting activity?

### Student Interpretation
Answer-01


Answer-02

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

# Count observations by reporting period
timeline = df["Time Period"].value_counts().sort_index()

# Create figure
plt.figure(figsize=(16, 6))

# Plot line
plt.plot(
    timeline.index,
    timeline.values,
    color="#1f77b4",
    linewidth=3,
    marker="o",
    markersize=7,
    markerfacecolor="#ff7f0e",
    markeredgecolor="black",
    markeredgewidth=0.8
)

# Highlight highest reporting period(s)
max_value = timeline.max()
max_periods = timeline[timeline == max_value]

plt.scatter(
    max_periods.index,
    max_periods.values,
    color="crimson",
    s=120,
    zorder=5,
    label="Highest Reporting Activity"
)

# Annotate highest reporting period(s)
for x, y in zip(max_periods.index, max_periods.values):
    plt.annotate(
        f"{y:,}",
        xy=(x, y),
        xytext=(0, 12),
        textcoords="offset points",
        ha="center",
        fontsize=10,
        fontweight="bold",
        color="crimson"
    )

# Titles
plt.title(
    "Quarterly Reporting Activity (2013Q4–2025Q4)",
    fontsize=18,
    fontweight="bold"
)

plt.xlabel("Reporting Period", fontsize=13)
plt.ylabel("Number of Observations", fontsize=13)

# Format y-axis
plt.gca().yaxis.set_major_formatter(
    mtick.StrMethodFormatter("{x:,.0f}")
)

# Show every second quarter for readability
plt.xticks(
    ticks=range(0, len(timeline.index), 2),
    labels=timeline.index[::2],
    rotation=45,
    ha="right",
    fontsize=10
)

# Grid
plt.grid(axis="y", linestyle="--", alpha=0.4)

# Remove top and right borders
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

# Legend
plt.legend(frameon=False, fontsize=11)

plt.tight_layout()
plt.show()
```

# Step 11: Correlation

Definition
> Correlation measures the strength and direction of the relationship between two numerical variables. The correlation coefficient ranges from −1 to +1.

Why are we doing it?
> Correlation analysis helps identify variables that move together, detect redundant variables, and understand relationships before performing statistical modelling or machine learning.

Formula
$$
r=\frac{\mathrm{Cov}(X,Y)}{\sigma_X\sigma_Y}
$$

Where:

- $r$ = Correlation coefficient
- $\mathrm{Cov}(X,Y)$ = Covariance between two variables
- $\sigma_X, \sigma_Y$ = Standard deviations of $X$ and $Y$
