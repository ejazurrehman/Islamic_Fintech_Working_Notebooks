### 5. Modulus Operator (%)
- The modulus operator returns the remainder after dividing one number by another.

```python
print(13%2)
```

### 6. Floor Division Operator (//)
- The floor division operator (//) divides one number by another and returns the quotient without the decimal part. It rounds the result down to the nearest whole number (toward negative infinity).

```python
print(8//4)
```

### 7. Exponentiation Operator (**)
- The exponentiation operator (**) raises a number (called the base) to the power of another number (called the exponent).

```python
print(2**3)
```

### 8. BODMAS
- it is a rule that determines the correct order of mathematical operations in an expression.

| Letter | Meaning        | Description                                                        |
| ------ | -------------- | ------------------------------------------------------------------ |
| **B**  | Brackets       | Solve expressions inside brackets first: `()`, `[]`, `{}`          |
| **O**  | Orders         | Evaluate powers (exponents), roots, and indices, e.g., `2³`, `√16` |
| **D**  | Division       | Perform division from left to right                                |
| **M**  | Multiplication | Perform multiplication from left to right                          |
| **A**  | Addition       | Perform addition from left to right                                |
| **S**  | Subtraction    | Perform subtraction from left to right                             |


```python
print(2 + 3 * 4)
```

### 9. PEMDAS
- It is another rule for the order of mathematical operations, primarily used in the United States. It is equivalent to BODMAS, which is commonly used in the UK, Pakistan, and many Commonwealth countries.

| Letter | Meaning        | Description                                          |
| ------ | -------------- | ---------------------------------------------------- |
| **P**  | Parentheses    | Solve expressions inside `()`, `[]`, and `{}` first. |
| **E**  | Exponents      | Evaluate powers (e.g., `2³`, `5²`) and roots.        |
| **M**  | Multiplication | Perform multiplication from left to right.           |
| **D**  | Division       | Perform division from left to right.                 |
| **A**  | Addition       | Perform addition from left to right.                 |
| **S**  | Subtraction    | Perform subtraction from left to right.              |


```python
print(3**2/2*3/3+6-4/(2*6))
```

# Conditional Logic in Python

Conditional logic is a programming technique that allows a program to make decisions based on whether a condition is True or False. In Python, conditional logic is implemented using if, elif, and else statements.

if condition:
    # Code executes if the condition is True

elif another_condition:
    # Code executes if the first condition is False
    # and this condition is True

else:
    # Code executes if all conditions are False

### Profit or Loss (Using if and else)

### Loan Approval System

```python
# Function to evaluate loan application
def loan_approval():

    # Input Variables
    applicant_name = input("Enter Applicant Name: ")
    monthly_income = float(input("Enter Monthly Income (Rs.): "))
    credit_score = int(input("Enter Credit Score: "))
    loan_amount = float(input("Enter Loan Amount (Rs.): "))

    print("\n----- Loan Application Summary -----")
    print("Applicant Name :", applicant_name)
    print("Monthly Income :", monthly_income)
    print("Credit Score   :", credit_score)
    print("Loan Amount    :", loan_amount)

    # Conditional Logic
    if monthly_income >= 100000 and credit_score >= 750:
        print("\nLoan Status : Approved")
        print("Reason : High income and excellent credit score.")

    elif monthly_income >= 70000 and credit_score >= 650:
        print("\nLoan Status : Conditional Approval")
        print("Reason : Requires additional documents.")

    elif monthly_income < 50000 or credit_score < 600:
        print("\nLoan Status : Rejected")
        print("Reason : Low income or poor credit score.")

    else:
        print("\nLoan Status : Under Review")
        print("Reason : Further financial assessment required.")

# Function Call
loan_approval()
```

### Investment Recommendation

```python
def investment_advisor():

    investor_name = input("Enter Investor Name: ")
    expected_return = float(input("Expected Return (%): "))
    risk_level = input("Risk Level (Low/Medium/High): ")

    print("\nInvestment Recommendation")

    if expected_return >= 15 and risk_level == "Low":
        print("Strong Buy")

    elif expected_return >= 10 and risk_level == "Medium":
        print("Buy")

    elif expected_return < 10 or risk_level == "High":
        print("Avoid Investment")

    else:
        print("Hold and Monitor")

investment_advisor()
```

### Company Performance Evaluation

```python
def company_performance():

    revenue = float(input("Enter Revenue (Rs.): "))
    expenses = float(input("Enter Expenses (Rs.): "))

    profit = revenue - expenses

    print("\nNet Profit =", profit)

    if profit > 1000000:
        print("Excellent Performance")

    elif profit > 500000:
        print("Good Performance")

    elif profit > 0:
        print("Average Performance")

    else:
        print("Company is in Loss")

company_performance()
```

### Shariah Compliance Screening

```python
# Function to evaluate Shariah compliance
def shariah_screening():

    # Input Variables
    company_name = input("Enter Company Name: ")
    debt_ratio = float(input("Enter Debt Ratio (%): "))
    non_halal_income = float(input("Enter Non-Halal Income (%): "))

    print("\n----- Shariah Screening Report -----")
    print("Company Name           :", company_name)
    print("Debt Ratio             :", debt_ratio, "%")
    print("Non-Halal Income Ratio :", non_halal_income, "%")

    # Decision Making
    if debt_ratio <= 33 and non_halal_income <= 5:
        print("\nStatus : Shariah-Compliant")
        print("Recommendation : Eligible for Islamic Investment.")

    elif debt_ratio <= 40 and non_halal_income <= 10:
        print("\nStatus : Requires Shariah Board Review")
        print("Recommendation : Further analysis is required.")

    else:
        print("\nStatus : Non-Compliant")
        print("Recommendation : Not suitable for Islamic Investment.")

# Function Call
shariah_screening()
```

# Loops in Python

Definition

A loop is a programming structure that repeatedly executes a block of code until a specified condition is met or for a specified number of iterations. Loops eliminate the need to write the same code multiple times.

Python provides two main types of loops:

1. for loop: Used when the number of repetitions is known.
2. while loop: Used when the number of repetitions depends on a condition.

1. for Loop

Definition: A for loop executes a block of code a fixed number of times or iterates over a sequence such as a list, range, or string.

Syntax:\
for variable in sequence:\
    # Code to execute

### Calculate Total Investment of 5 Investors

```python
total_investment = 0

for i in range(1, 6):
    investment = float(input(f"Enter Investment of Investor {i} (Rs.): "))
    total_investment += investment

print("\nTotal Investment = Rs.", total_investment)
```

### Calculate Annual Profits

```python
profits = [120000, 145000, 98000, 175000]

for profit in profits:
    print("Annual Profit = Rs.", profit)
```

### Calculate Zakat for Multiple Customers

```python
nisab = 180000

for i in range(1, 6):

    savings = float(input(f"Customer {i} Savings (Rs.): "))

    if savings >= nisab:
        zakat = savings * 0.025
        print("Zakat Payable = Rs.", zakat)

    else:
        print("Not Eligible for Zakat")
```

### Screen Multiple Companies

```python
for i in range(1, 4):

    company = input("Company Name: ")
    debt_ratio = float(input("Debt Ratio (%): "))

    if debt_ratio <= 33:
        print(company, "is Shariah-Compliant")

    else:
        print(company, "is Non-Compliant")
```

2. while Loop

Definition: A while loop repeatedly executes a block of code as long as a specified condition remains True.

Syntax:\
while condition:\
    # Code

### Deposit Until Target is Achieved

```python
target = 500000
balance = 0

while balance < target:

    deposit = float(input("Enter Deposit Amount: "))
    balance += deposit

print("\nInvestment Target Achieved!")
print("Total Balance = Rs.", balance)
```

### Continue Collecting Zakat Until User Stops

```python
total_zakat = 0

choice = "yes"

while choice == "yes":

    zakat = float(input("Enter Zakat Amount: "))
    total_zakat += zakat

    choice = input("Add another payment? (yes/no): ")

print("\nTotal Zakat Collected = Rs.", total_zakat)
```

### Using a Function and a Loop

```python
def calculate_dividend():

    total_dividend = 0

    for i in range(1, 4):

        dividend = float(input(f"Enter Dividend of Company {i} (Rs.): "))
        total_dividend += dividend

    print("\nTotal Dividend Received = Rs.", total_dividend)

calculate_dividend()
```
