```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Count observations by reporting period
timeline = df["Time Period"].value_counts().sort_index().reset_index()
timeline.columns = ["Time Period", "Observations"]

# Extract year and quarter
timeline["Year"] = timeline["Time Period"].astype(str).str[:4]
timeline["Quarter"] = timeline["Time Period"].astype(str).str[-2:]

# Create year-quarter heatmap table
quarter_heatmap = timeline.pivot(
    index="Year",
    columns="Quarter",
    values="Observations"
)

# Keep quarters in correct order
quarter_heatmap = quarter_heatmap.reindex(columns=["Q1", "Q2", "Q3", "Q4"])

# Create figure
plt.figure(figsize=(10, 7))

ax = sns.heatmap(
    quarter_heatmap,
    annot=True,
    fmt=".0f",
    cmap="YlGnBu",
    linewidths=1.5,
    linecolor="white",
    cbar_kws={"label": "Number of Observations"}
)

plt.title(
    "Islamic Banking Reporting Activity by Year and Quarter",
    fontsize=17,
    fontweight="bold",
    pad=18
)

plt.xlabel("Quarter", fontsize=12, fontweight="bold")
plt.ylabel("Year", fontsize=12, fontweight="bold")

plt.xticks(fontsize=11, fontweight="bold")
plt.yticks(fontsize=11, fontweight="bold", rotation=0)

plt.tight_layout()
plt.show()

# Show table below graph
quarter_heatmap
```

Understanding the Coefficient of Variation (CV)
> The Coefficient of Variation (CV) measures the relative variability of a variable by comparing its standard deviation to its mean. A higher CV indicates greater variability relative to the average value.

Formula:

$$
\text{Coefficient of Variation (CV)} = \frac{\text{Standard Deviation}}{\text{Mean}}
$$

# Step 11: Correlation

Definition
> Correlation measures the strength and direction of the relationship between two numerical variables. The correlation coefficient ranges from −1 to +1.

Why are we doing it?
> Correlation analysis helps identify variables that move together, detect redundant variables, and understand relationships before performing statistical modelling or machine learning.

Formula
$$
r=\frac{\mathrm{Cov}(X,Y)}{\sigma_X\sigma_Y}
$$

Where:

- $r$ = Correlation coefficient
- $\mathrm{Cov}(X,Y)$ = Covariance between two variables
- $\sigma_X, \sigma_Y$ = Standard deviations of $X$ and $Y$

## Research Questions

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Correlation matrix
corr = df.select_dtypes(include="number").corr()

# Short labels
labels = ["USD Rate", "Values", "USD Millions"]

plt.figure(figsize=(7,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="RdYlBu_r",
    fmt=".2f",
    square=True,
    linewidths=0.5,
    xticklabels=labels,
    yticklabels=labels,
    cbar_kws={"shrink":0.8},
    annot_kws={"fontsize":11}
)

plt.title(
    "Correlation Matrix",
    fontsize=16,
    fontweight="bold"
)

plt.xticks(rotation=0, fontsize=11)
plt.yticks(rotation=0, fontsize=11)

plt.tight_layout()
plt.show()
```
